```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head(50)
```

```python
Avec la commande ci-dessous nous pouvons résoudre les problèmes NaN avec la principale méthode de remplissage de valeurs nan
cette méthode consiste tout simplement à remplacer le valeur NaN par le valeur le plus fréquent dans la colonne
```

```python
df["tax"] = df["tax"].fillna(df["tax"].mode()[0]) # Méthode adaptée pour la colonne tax
df["country"] = df["country"].fillna(df["country"].mode()[0]) # Méthode adaptée pour la colonne country
df["gender"] = df["gender"].fillna(df["gender"].mode()[0])# Méthode adaptée pour la colonne gender
```
