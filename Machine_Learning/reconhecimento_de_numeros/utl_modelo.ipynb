# Test du modèle de reconnaissance de chiffres

```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
```

```python

# Charger le modèle
model = load_model("digit_model.h5")
```

```python

# Charger l'image (adapter le chemin si nécessaire)
img = cv2.imread("2.jpg", cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img, (28, 28))
img = 255 - img  # inversion (fond noir MNIST)
img = img / 255.0
```

```python

# Reshape pour le modèle
img_array = img.reshape(1, 28, 28, 1)
```

```python

# Prédiction
prediction = model.predict(img_array)
chiffre = np.argmax(prediction)
confidence = np.max(prediction) * 100

print(f"Chiffre reconnu : {chiffre}")
print(f"Confiance : {confidence:.2f}%")
```

```python

# Affichage
plt.imshow(img, cmap="gray")
plt.title(f"Chiffre reconnu : {chiffre}")
plt.axis("off")
plt.show()
```
